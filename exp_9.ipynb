```python




import numpy as np

import matplotlib.pyplot as plt

from sklearn.cluster import KMeans

from sklearn.datasets import make_blobs



# Generating synthetic data using make_blobs

n_samples = 300

centers = 4  # Number of clusters

X, _ = make_blobs(n_samples=n_samples, centers=centers, random_state=42)



# Applying KMeans clustering

kmeans = KMeans(n_clusters=centers)

kmeans.fit(X)



# Getting cluster labels and cluster centers

cluster_labels = kmeans.labels_

cluster_centers = kmeans.cluster_centers_



# Plotting the clusters

plt.figure(figsize=(8, 6))



# Plotting data points with colors representing clusters

plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='viridis', edgecolor='k')



# Plotting cluster centers

plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='o', s=200, c='red', label='Cluster Centers')

plt.title('K-means Clustering')

plt.legend()

plt.xlabel('Feature 1')

plt.ylabel('Feature 2')

plt.grid(True)

plt.show()
```
